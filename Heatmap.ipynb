```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
oly_file = pd.read_csv("C:/Users/aakan/Desktop/Stevens/Semester 2/Self_Study/Python/Pandas/Olympics/Python_Project/olympics.csv",skiprows=4)
oly_file
```

```python
o1 = oly_file[oly_file.Edition == 2008].groupby(['NOC','Medal']).size().unstack('Medal',fill_value = 0).sort_values(['Gold','Silver','Bronze'],ascending=False)[['Gold','Silver','Bronze']]
o1
```

```python
o1t = o1.transpose()
o1t
```

```python
sns.heatmap(o1t)
```

```python
cmap = sns.cm.rocket_r
```

```python
plt.figure(figsize=(15,5))
sns.heatmap(o1t,cmap=cmap)
```
