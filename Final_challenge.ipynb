```python
import pandas as pd
```

```python
oly_file = pd.read_csv("C:/Users/aakan/Desktop/Stevens/Semester 2/Self_Study/Python/Pandas/Olympics/Python_Project/olympics.csv",skiprows=4)
oly_file
```

```python
a=oly_file[oly_file.NOC == 'USA'].groupby(['Edition','Athlete','Medal']).size().unstack('Medal',fill_value=0)
```

```python
a
```

```python
a['total_medals'] = a['Gold']+a['Silver']+a['Bronze']
```

```python
a.reset_index(inplace=True)
```

```python
a
```

```python
tu = [group.sort_values('total_medals',ascending =False)[:1] for year,group in a.groupby('Edition')]
tu
    
   
```

```python
new_df = pd.DataFrame()
```

```python
for i in tu:
    new_df = new_df.append(i)
```
